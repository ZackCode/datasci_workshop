# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### Evaluate the Claims in the report

In report there are four claims:

1.Overall, rate of readmissions is trending down with increasing number of discharges

2.With lower number of discharges, there is a greater incidence of excess rate of readmissions (area shaded red), With higher number of discharges, there is a greater incidence of lower rates of readmissions (area shaded green

3.In hospitals/facilities with number of discharges > 1000, mean excess readmission rate is 0.978 and 44% have excess readmission rate greater than 1

4.In hospitals/facilities with number of discharges < 100, mean excess readmission rate is 1.023 and 63% have excess readmission rate greater than 1

In order to test these claims, I divide the data into four groups: hospitals with discharge number under 100, ones with discharge number between 100 and 300, ones with discharge number between 300 and 1000, and ones with discharge number above 1000.

Two null hypothesis:

1.There is no difference between the readmission rate of hospitals with discharge number under 100 and the readmission rate of hospitals with discharge number over 1000.

2.There is no difference between the readmission rate of hospitals with discharge number between 100 and 300 and the readmission rate of hospitals with discharge number between 300 and 1000.

The means of these groups as well as 99% confidence interval is calculated, as well as p values and 99% interval for null hypothesis.

```python
df = clean_hospital_read_df[81:-3][['Number of Discharges','Excess Readmission Ratio']]

df_lower100 = df[df['Number of Discharges']< 100]
df_bet100and300 = df[(df['Number of Discharges']>= 100) & (df['Number of Discharges']< 300)]
df_bet300and1000 = df[(df['Number of Discharges']>= 300) & (df['Number of Discharges']< 1000)]
df_above1000 = df[df['Number of Discharges']>= 1000]

#print df_lower100.head()
#print df_bet100and300.head()
#print df_bet300and1000.head()
#print df_above1000.head()

def bsfromfunc(observes,targetfunc):
    value = targetfunc(observes)
    bs_target = np.empty(100000)
    for i in range(100000):
        bs_sample = np.random.choice(observes,size=len(observes))
        bs_target[i] = targetfunc(bs_sample)
        
    print 'value, ', value
    print '99% interval, ', np.percentile(bs_target, [0.5, 99.5])
    
    return value,np.percentile(bs_target, [0.5, 99.5])
    
print bsfromfunc(df_lower100['Excess Readmission Ratio'],np.mean)
print bsfromfunc(df_bet100and300['Excess Readmission Ratio'],np.mean)
print bsfromfunc(df_bet300and1000['Excess Readmission Ratio'],np.mean)
print bsfromfunc(df_above1000['Excess Readmission Ratio'],np.mean)
```

```python
def perm_diff(data1,data2,targetfun):
    conc_data = np.concatenate((data1,data2))
    value = targetfun(data1) - targetfun(data2)
    value_diff = np.empty(100000)

    for i in range(100000):
        perm_data = np.random.permutation(conc_data)
        perm_data1 = perm_data[:len(data1)]
        perm_data2 = perm_data[len(data1):]
        
        perm_value1 = targetfun(perm_data1)
        perm_value2 = targetfun(perm_data2)
        value_diff[i] = perm_value1 - perm_value2
        
        if value>0:
            p = np.sum(value_diff > value)/len(value_diff)
        else:
            p = np.sum(value_diff < value)/len(value_diff)
    print "p value:",  p
    print "value", value
    print "99% null hypothesis interval:",  np.percentile(value_diff, [0.5, 99.5])

print "\n null hypothese: there is no difference for readmission rate between hospitals with discharge under 100 and discharge above 1000"
perm_diff(df_lower100['Excess Readmission Ratio'],df_above1000['Excess Readmission Ratio'],np.mean)

print "\n null hypothese: there is no difference for readmission rate between hospitals with discharge 100 to 300 and discharge 300 to 1000"
perm_diff(df_bet100and300['Excess Readmission Ratio'],df_bet300and1000['Excess Readmission Ratio'],np.mean)
                           
```

From above results, we can conclude:
    
Both null hypothesizes are not true, therefore with discharge number going high, the readmission rate is indeed lower.
